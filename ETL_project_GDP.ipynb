creating an automated script that can extract the list of all countries in order of their GDPs in billion USDs, as logged by the International Monetary Fund (IMF). Since IMF releases this evaluation twice a year, this code will be used by the organization to extract the information as it is updated.

```python
from bs4 import BeautifulSoup
import requests
```

```python
html=requests.get("https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29").text
soup= BeautifulSoup(html,'html.parser')
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns=['Country','GDP_USD_million'])
data
```

```python
dict_list=[]
for row in rows:
    col = row.find_all('td')
    # print(col)
    if len(col)!=0 and col[0].a!=None:    # header excl. and world row also exclude
      country_name = col[0].a.string
      IMF_estimate = col[2].string
      row_dict={
          'Country':country_name,
          'GDP_USD_million': IMF_estimate,
      }

      dict_list.append(row_dict)

print(dict_list)
```

```python
data=pd.DataFrame.from_dict(dict_list)
print(data)
```

# display only the entries with more than a 100 billion USD economy

```python
# create another dataframe
final_data = data.copy()
final_data.rename(columns = {'GDP_USD_million':'GDP_USD_billion'}, inplace = True)
print(final_data)
```

```python
# Data cleaning required
final_data['GDP_USD_billion']= final_data['GDP_USD_billion'].str.replace(',', '')
print(final_data)
```

```python
# need to clean code as few rows have "—" in gdp
import numpy as np
# replace "—" to NaN
# final_data.replace("—", np.nan, inplace = True)

final_data.replace("—", 0, inplace = True)
final_data.isnull().sum()
```

```python
final_data['GDP_USD_billion']= final_data['GDP_USD_billion'].astype(int)
print(final_data)
```

```python
# million to billion conversion
final_data['GDP_USD_billion']= (final_data['GDP_USD_billion']/1000).round(2)
print(final_data)
```
